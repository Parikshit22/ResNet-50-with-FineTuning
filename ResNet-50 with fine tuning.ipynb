```python
data_path = r"C:\Users\MUJ\CodingBlocks\Deep Learning- CNN\dataset"
```

```python
import os
import numpy as np
```

```python
folders = os.listdir(data_path)
label_dict = {'cats':0,'dogs':1,'horses':2,'humans':3}
image_data= []
label = []
```

```python
from keras.preprocessing import image
for i in folders:
    path0 = os.path.join(data_path,i)
    for j in os.listdir(path0):
        img = os.path.join(path0,j)
        img = image.load_img(img,target_size = (224,224))
        img = image.img_to_array(img)/255.0
        image_data.append(img)
        label.append(label_dict[i])
```

```python
support = list(zip(image_data,label))
```

```python
import random
random.shuffle(support)
image_data[:],label[:] = zip(*support) 
```

```python
image_data = np.asarray(image_data)
label = np.asarray(label)
```

```python
from keras.utils import np_utils
label = np_utils.to_categorical(label)
```

```python
print(image_data.shape,label.shape)
```

```python
from keras.applications.resnet50 import ResNet50
from keras.layers import *
from keras.models import Model
```

```python
model = ResNet50(include_top = False,input_shape = (224,224,3),weights = 'imagenet')
model.summary()
```

```python
av1 = GlobalAveragePooling2D()(model.output)
fc1 = Dense(256,activation = 'relu')(av1)
d1 = Dropout(0.5)(fc1)
fc2 = Dense(4,activation = 'relu')(d1)

new_model = Model(input = model.input,outputs = fc2)
new_model.summary()
```

```python
for i in range(len(new_model.layers)):
    print(i,new_model.layers[i])
```

```python
for i in range(169):
    new_model.layers[i].trainable = False
```

```python
new_model.compile(optimizer="adam",loss = "binary_crossentropy",metrics=["acc"])
```

```python
new_model.fit(image_data,label,epochs=4,batch_size=16,validation_split=0.2)
```
